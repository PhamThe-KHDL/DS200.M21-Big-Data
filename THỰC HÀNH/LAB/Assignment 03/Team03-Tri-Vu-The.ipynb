```python
import pyspark
from pyspark.sql import SparkSession
from heapq import nlargest, nsmallest
```

## **Câu 1:** For each month, show the number of distinct players, and the total cost, the results are sorted by moth

```python
rdd_totalCost = trans240RDD.map(lambda x: x.split(',')).map(lambda x: (x[2],x[3])) \
.reduceByKey(lambda f1,f2: float(f1)+float(f2)).sortByKey()
rdd_totalCost.collect()
```

```python
rdd_month = trans240RDD.map(lambda x: x.split(',')).map(lambda x: (x[1], x[2])) \
.map(lambda x: (x[0].split('-'), x[1])).map(lambda x: (x[0][0], (x[1]))) \
.distinct().map(lambda x: (x[1], x[0])).sortBy(lambda x: x[1])
rdd_month.collect()
```

```python
rdd_month_totalCost = rdd_totalCost.join(rdd_month).map(lambda x: (x[1][1], x[0], x[1][0], 1))
rdd_month_totalCost.sortBy(lambda x: x[0]).collect()
```

```python
trans240RDD.map(lambda x: x.split(',')).map(lambda x: (x[1], x[2], 1)) \
.map(lambda x: (x[0].split('-'), x[1], x[2])).map(lambda x: (x[0][0], (x[1], x[2]))) \
.distinct().reduceByKey(lambda f1, f2: (f1[0]+','+f2[0], f1[1]+f2[1])).sortByKey() \
.map(lambda x: (x[0], x[1][0], x[1][1])).collect()
```

```python
rdd_month_totalCost.map(lambda x: (x[0], (x[2], x[3]))) \
.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1])).sortBy(lambda x: x[0]) \
.map(lambda x: (x[0], x[1][1], x[1][0])).collect()
```

## **Câu 5:** For each state, show the name of three oldest player in each state, the results are sorted by the state

```python
from heapq import nlargest
trans240RDD.map(lambda x: x.split(",")).map(lambda x: (x[2], x[7])).distinct() \
.join(custRDD.map(lambda x: x.split(",")).map(lambda x: (x[0], (x[1], x[3])))) \
.map(lambda x: (x[1][0], (x[1][1][0], x[1][1][1]))) \
.groupBy(lambda x: x[0]).flatMap(lambda g: nlargest(3, g[1], key=lambda x: x[1][1])) \
.map(lambda x: (x[0], x[1][0])).reduceByKey(lambda x,y: x + " - " + y).sortByKey().collect()
```

## **Câu 6:** For each state, show the average age of players in each state, the results are sorted by the average age

## **Câu 13 :** For each player ID, show the list of three games with most transactions, and the list of the number of transactions of these three games, the results are sorted by player IDs
